### Задача 1. Несмещенность оценок.

**Теория**

Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathsf{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно. 
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

a). Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$ и является состоятельной оценкой $\sigma^2$;

b). Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$?

c*). Докажите, что если конечны первые четыре момента ( $\mathsf{E}X, \mathsf{E}X^2, \mathsf{E}X^3, \mathsf{E}X^4$) распределения  $\mathsf{P}$, то статистика $S^2$ является асимптотически нормальной оценкой $\sigma^2$ и найдите ее асимптотическую дисперсию.
        

**Ответ:**
**a)**
Воспользуемся формулой $\overline{X} = \frac{1}{n}\sum\limits_{i=1}^nX_i$  
$\frac1n\sum\limits_{i=1}^n\left(X_i-\overline{x}\right)^2 = \frac1n\sum\limits_{i=1}^n\left(X_i^2 - 2X_i\overline{X}+\overline{X}^2\right) = \frac{1}{n}\sum\limits_{i=1}^n\left(X_i^2\right) - \frac{2}{n}\sum\limits_{i=1}^n\left(X_i\overline{X}\right)+\frac{1}{n}\sum\limits_{i=1}^n\left(\overline{X}^2\right) = \overline{X^2}  - \frac{2}{n}\sum\limits_{i=1}^n\left(X_i\overline{X}\right) + \overline{X}^2 = $   
$ = \overline{X^2} - 2\overline{X}\frac{1}{n}\sum\limits_{i=1}^n\left(X_i\right)+ \overline{X}^2 = \overline{X^2} - 2\overline{X}^2 + \overline{X}^2 = \overline{X^2} - \overline{X}^2$   
Ч.Т.Д. т.к. $\overline{X^2}-\overline{X}^2 = S^2$  
**b)**
Представим  $S^2=\overline{X^2} - \overline{X}^2$ в виде:   
$\left(\frac{1}{n}\sum\limits_{i=1}^nX_i^2\right) - \left(\frac{1}{n}\sum\limits_{i=1}^nX_i\right)^2 = \left(\frac{1}{n}\sum\limits_{i=1}^nX_i^2\right) - \left(\frac{1}{n^2}\sum\limits_{i=1}^nX_i^2\right) - \left(\frac{2}{n^2}\sum\limits_{j\neq i}X_iX_j\right) = \left(\frac{n - 1}{n^2}\sum\limits_{i=1}^nX_i^2\right) - \left(\frac{2}{n^2}\sum\limits_{j\neq i}X_iX_j\right)$  
Найдём мат ожидание от этого представление $S^2$:  
${\sf E}S^2 = \left(\frac{n - 1}{n^2}\sum\limits_{i=1}^n{\sf E}(X_i^2)\right) - \left(\frac{2}{n^2}\sum\limits_{j\neq i}{\sf E}(X_iX_j)\right)$.$(1)$  
Так как оценка будет не смещенной еслит условие выполняется при любом параметре, то рассмотрим случай $E\left(X_i\right)=0$, тогда:  
1) Из ${\sf D}X_i = {\sf E}(X_i^2) - {\sf E}(X_i)^2$ следует, что при ${\sf E}X_i = 0$ получим ${\sf D}X_i = {\sf E}(X_i ^ 2)$  
2) Т.к. $X_1,...,X_n$ - выборка, то при $i \neq j$ $X_i$ и $X_j$ являются независимыми, тогда ${\sf E}(X_iX_j) = {\sf E}X_i{\sf E}X_j = 0$ при ${\sf E}X_i = 0$  
Значит, применив это всё к формуле $(1)$ получим:  
$\frac{n-1}{n^2}\sum\limits_{i=1}^n{\sf D}X_i = \frac{n - 1}{n^2}n{\sf D}X_i = \frac{n - 1}{n}{\sf D}X_i = \frac{n - 1}{n}\sigma^2$  
${\sf E}S^2 = \frac{n - 1}{n}{\sf D}X_i \neq {\sf D}X_i$  
Значит оценка не является несмещенной


**Практика** 

Пусть теперь $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$.

**Замечание:** если некоторые оценки имеют большой разброс, и разница между графиками зависимостей оценок с малыми значениями недостаточно заметна, стоит сделать два графика, на одном из которых будут изображены все оценки, а на втором &mdash; только достаточно хорошие (мало отличающиеся от оцениваемого параметра), то есть этот график будует иметь меньший масштаб по оси $y$.

**Решение:**

```
sample = sps.cauchy.rvs(size=1000)
n = np.linspace(1, 1000, 1000)

avg = np.cumsum(sample) / n
median = [np.median(sample[0:i+1]) for i in range(1000)]
one_step = np.zeros(1000)
for i in range(1000):
  one_step[i] = median[i] + np.sum((sample[0:i+1] - median[i]) / (1 + (sample[0:i+1] - median[i])**2)) / np.sum((1 - (sample[0:i+1] - median[i])**2) / (1 + (sample[0:i+1]
  - median[i])**2)**2)

plt.figure(figsize=(10,5))
plt.plot(n, avg, label = 'выб. сред.', color = 'orange')
plt.plot(n, median, label = 'медиана', color = 'red')
plt.plot(n, one_step, label = 'одношаг. оценка', color = 'blue')
plt.legend()
plt.show()

plt.plot(n, median, label = 'медиана', color = 'red')
plt.plot(n, one_step, label = 'одношаг. оценка', color = 'blue')
plt.legend()
plt.show()
```
